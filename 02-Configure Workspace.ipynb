```python
import azureml.core
import os

from azureml.data.data_reference import DataReference
from azureml.data.datapath import DataPath
from azureml.core import Workspace, Datastore, Dataset
from dotenv import load_dotenv
load_dotenv()
```

```python
ws = Workspace.from_config('.azureml/config.json')
def_blob_store = Datastore(ws, "workspaceblobstore")
```

```python
def_blob_store.upload(
    src_dir="./train_dataset",
    target_path="fulltrain",
    overwrite=True)
```

```python
datastore_paths = [(def_blob_store, 'fulltrain/train')]
dataset_img = Dataset.File.from_files(datastore_paths)
dataset_img = dataset_img.register(ws, name='fulltrain_img', description='Fulltrain image dataset', tags={'resize': '50%'}, create_new_version=False)

datastore_paths = [(def_blob_store, 'fulltrain/train_labels.csv')]
dataset_labels = Dataset.Tabular.from_delimited_files(path=datastore_paths)
dataset_labels = dataset_labels.register(ws, name='fulltrain_labels', description='Fulltrain labels dataset', create_new_version=False)
```
